```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('/config/workspace/Dataset/diabetes.csv')
df.head()
```

```python
df.describe().T
```

* Outcome:-
Except Pregnacy and outcome feature other features cannot have 0 value so we need to replace it by there respective central measures.

```python
df['BMI']=df['BMI'].replace(0,df['BMI'].mean())
df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].median())
df['SkinThickness']=df['SkinThickness'].replace(0,df['SkinThickness'].median())
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].median())
df['DiabetesPedigreeFunction']=df['DiabetesPedigreeFunction'].replace(0,df['DiabetesPedigreeFunction'].mean())
df['Age']=df['Age'].replace(0,df['Age'].median())
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
import pickle
def StandardScaler1(x_train,x_test):
    scaler=StandardScaler()
    x_train_scaled=scaler.fit_transform(x_train)
    x_test_scaled=scaler.transform(x_test)
    pickle.dump(scaler,open('/config/workspace/Dataset/Scaled.pkl','wb'))

    return x_train_scaled,x_test_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
StandardScaler1(x_train,x_test)
```

```python
# Descision Tree


parameter_tree={'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']}  

grid_tree=GridSearchCV(DecisionTreeClassifier(),param_grid=parameter_tree,refit=True,cv=3,verbose=3,scoring='accuracy')
grid_tree.fit(x_train,y_train)
```

```python
grid_tree.best_params_
```

```python
tree_classifier=DecisionTreeClassifier(criterion= 'gini',
 max_depth= 3,
 max_features= 'log2',
 splitter= 'best')
tree_classifier.fit(x_train,y_train)
y_pred_tree=tree_classifier.predict(x_test)
print(accuracy_score(y_test,y_pred_tree))
print(confusion_matrix(y_test,y_pred_tree))
print(classification_report(y_test,y_pred_tree))
```
